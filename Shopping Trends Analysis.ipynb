## 2. How does the average purchase amount vary across different product categories?

```python
shop['Category'].unique()
```

```python
category_means= shop.groupby('Category')['Purchase Amount (USD)'].mean().to_dict()
```

```python
# Extract categories and mean values
categories = list(category_means.keys())
means = list(category_means.values())

# bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, means, color='skyblue')

plt.xlabel('Category')
plt.ylabel('Mean Purchase Amount (USD)')
plt.title('Mean Purchase Amount by Category')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

## Observations:

- Footwear has the highest mean purchase amount, followed closely by Clothing and Accessories.
- Outerwear has the lowest mean purchase amount among the four categories.

## 4. What are the most commonly purchased items in each category?

```python
shop.columns
```

```python
top_items_per_category= shop.groupby('Category')['Item Purchased'].value_counts()
top_items_per_category
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Top 5 Purchased Items per Category', fontsize=16)

categories = top_items_per_category.index.get_level_values(0).unique()

for i, category in enumerate(categories):
    ax = axes[i // 2, i % 2]
    ax.pie(top_items_per_category[category], labels=top_items_per_category[category].index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Top 5 in {category}')

plt.tight_layout()
plt.show()
```

## 11. Are there any correlations between the size of the product and the purchase amount?

```python

shop_group = shop.groupby(['Size', 'Category'])['Purchase Amount (USD)'].sum().reset_index()

# Pivot the table to make 'Size' the index and 'Category' the columns
pivot_table = shop_group.pivot(index='Size', columns='Category', values='Purchase Amount (USD)')

# Create heatmap
sns.heatmap(pivot_table, cmap='Reds', annot=True, fmt=".2f")

plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Purchase Amount Heatmap by Size and Category')

plt.show()
```



## Observations:

**Overall Purchase Trends:**

- Clothing consistently shows the highest purchase amounts across all sizes, indicating it's the most popular category.
- Accessories also have relatively high purchase amounts across sizes, suggesting strong demand.
- Footwear and Outerwear generally have lower purchase amounts compared to Clothing and Accessories.

**Size-Specific Trends:**

- Size "M" has the highest total purchase amount across all categories.
- Size "L" also shows significant purchase amounts, particularly in Clothing and Accessories.
- Size "S" and "XL" have lower overall purchase amounts compared to "M" and "L".
  
**Category-Specific Trends:**

- Clothing consistently has the highest purchase amounts across all sizes.
- Accessories show a strong presence across all sizes, although slightly lower than Clothing.
- Footwear and Outerwear have lower purchase amounts across all sizes compared to Clothing and Accessories.
